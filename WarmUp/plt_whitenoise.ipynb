```python
import numpy as np
import matplotlib.pyplot as plt

# matplotlibをjupyter内で表示させるおまじない
%matplotlib inline

from IPython.display import Audio # jupyter上で音声を再生するためのおまじない
```

```python
# 白色雑音のサンプル数を設定
n_samples = 16000

# サンプリング周波数
sample_freq = 16000
```

```python
# 白色雑音を生成
data = np.random.normal(size=n_samples)
```

```python
# 時間軸の設定
time = np.arange(n_samples) / sample_freq

# 音声データのプロット
plt.plot(time, data)

# x軸のラベル
plt.xlabel("Time (sec)")

# y軸のラベル
plt.ylabel("Amplitude")

# 画像のタイトル
plt.title("Waveform")
```

```python
# 白色雑音を再生（音量注意！）
Audio(data, rate=sample_freq) 
```
