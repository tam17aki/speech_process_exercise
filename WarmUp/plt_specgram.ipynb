```python
import numpy as np
from scipy.io import wavfile
import matplotlib.pyplot as plt

# matplotlibをjupyter内で表示させるおまじない
%matplotlib inline
```

```python
IN_WAVE_FILE = "in.wav"  # モノラル音声（前提）
```

```python
FRAME_LENGTH = 1024  # フレーム長
HOP_LENGTH = 80  # フレームのシフト長
FFT_LENGTH = FRAME_LENGTH  # FFTサイズ
N_OVERLAP = FRAME_LENGTH - HOP_LENGTH  # オーバーラップ幅
```

```python
# 音声データ読み込み (fsがサンプリング周波数、dataは音声データ)
fs, data = wavfile.read(IN_WAVE_FILE)
```

```python
# プロット枠を確保
plt.figure(figsize=(10, 4))

# x軸のラベル
plt.xlabel("Time (sec)")

# y軸のラベル
plt.ylabel("Frequency (Hz)")

# 画像のタイトル
plt.title("Spectrogram")

# スペクトログラムの表示
specgram = plt.specgram(data, NFFT=FFT_LENGTH, noverlap=N_OVERLAP, Fs=fs, cmap="jet")
```
