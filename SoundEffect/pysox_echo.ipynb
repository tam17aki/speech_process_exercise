```python
import sox
from scipy.io import wavfile
```

```python
from IPython.display import Audio
```

```python
IN_WAVE_FILE = "in.wav"       # 入力音声
```

```python
# トランスフォーマーをつくる
transformer = sox.Transformer()
```

```python
# エコー の パラメタ
n_echos = 2     # エコー回数
delays = [375]  # 遅延時間 (ms)
decays = [0.5]  # 減衰率
```

```python
# エコー回数分、遅延時間と減衰率を与える必要がある
# → エコー回数に等しい長さの「リスト」を 遅延時間と減衰率それぞれで用意する
# → n_echos が 2 なら遅延時間は [375, 750], 減衰率は [0.5, 0.25]
for i in range(1, n_echos):
    delays.append(delays[0] * (i + 1))   # 遅延時間は線形的
    decays.append(decays[0] ** (i + 1))  # 減衰率は指数的
```

```python
# エコーをかける用意
transformer.echo(n_echos=n_echos, delays=delays, decays=decays)
```

```python
# 音声ファイルを開く
sr, data = wavfile.read(IN_WAVE_FILE)
```

```python
# エコーをかける
echos = transformer.build_array(input_array=data, sample_rate_in=sr)
```

```python
Audio(data, rate=sr)  #エコー前
```

```python
Audio(echos, rate=sr)  #エコー後
```
