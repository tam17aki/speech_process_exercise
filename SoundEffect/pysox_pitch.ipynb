```python
import sox
from scipy.io import wavfile
```

```python
from IPython.display import Audio
```

```python
IN_WAVE_FILE = "in.wav"       # 入力音声
```

```python
# 音声ファイルを開く
sr, data = wavfile.read(IN_WAVE_FILE)
```

```python
# トランスフォーマーをつくる（単一音声に対する処理）
transformer = sox.Transformer()
```

```python
# ピッチシフト の パラメタ
# 単位：セミトーン（いわゆる半音 -> 1半音の変化は周波数的には約1.06倍）
# 正値は上げる、負値は下げる
# 実際にはfloat値を指定可能
PITCHSHIFT_HIGH = 3.0  # 3半音上げる
PITCHSHIFT_LOW = -5.0  # 5半音下げる
```

```python
# ピッチシフトをかける準備
transformer.pitch(n_semitones=PITCHSHIFT_HIGH)  # 上げる
# ピッチシフトをかける
pitch_high_data =transformer.build_array(input_array=data, sample_rate_in=sr)
```

```python
# ピッチシフトをかける準備
transformer.pitch(n_semitones=PITCHSHIFT_LOW)   # 下げる
# ピッチシフトをかける
pitch_low_data =transformer.build_array(input_array=data, sample_rate_in=sr)
```

```python
Audio(data, rate=sr)  #オリジナルの音声
```

```python
Audio(pitch_high_data, rate=sr)  #ピッチを上げた音声
```

```python
Audio(pitch_low_data, rate=sr)  #ピッチを下げた音声
```
