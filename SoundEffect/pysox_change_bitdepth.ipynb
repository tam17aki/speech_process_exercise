```python
import sox
from scipy.io import wavfile
from IPython.display import Audio
```

```python
IN_WAVE_FILE = "in.wav"       # 入力音声
OUT_WAVE_FILE = "out.wav"          # モノラル音声
```

```python
BITDEPTH = 8  # 8 bit
```

```python
# トランスフォーマーをつくる
transformer = sox.Transformer()
```

```python
# 量子化ビット数を8bitに変更
transformer.convert(bitdepth=BITDEPTH)
transformer.build(IN_WAVE_FILE, OUT_WAVE_FILE)
```

```python
# 音声ファイルを開く
sr, data = wavfile.read(IN_WAVE_FILE)
```

```python
# 量子化ビット数を8bitに変更
data_8bit = transformer.build_array(input_array=data, sample_rate_in=sr)
```

```python
Audio(data, rate=sr)  # 量子化bit数 変更前
```

```python
Audio(data_8bit, rate=sr)  # 量子化bit数 変更後 （量子化ノイズの発生！）
```
